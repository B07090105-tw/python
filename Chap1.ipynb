## 運算式

### 1.變數交換

```python
x=1
y=10

z=x
x=y
y=z

print(x)
print(y)
```

```python
x=1
y=10

x=x+y
y=x-y
x=x-y

print(x)
print(y)
```

## 變數
### 1.變數平行設定

```python
x,y,z=3,4,5

print(x)
print(y)
print(z)
```

```python
x,y,z=3,4,5
x += 1 # x=x+1
y *= 2 # y=y*2
z **= 3 # z=z^3

print(x,y,z)
```

### 2.變數的應用
Q:邊長為3,4,5的三角形.其面積大小為何?<br>
s=(a+b+c)/2<br>
sqrt(s(s-a)(s-b)(s-c))

```python
import math
a,b,c=3,4,5
s=(a+b+c)/2
d=math.sqrt(s*(s-a)*(s-b)*(s-c))

print(d)
```

Q:邊長為10的圓形.其面積大小為何?

```python
import math
r=10
print(r*r*math.pi)
```

## 型態轉換

### 1.不同型態之間所謂到的結果跟想像不同

```python
x="must"
print(10*x)
print("10"+x)
print(10,x)
```

### 2.強制型別轉換

```python
x=12
print(type(x))
y=float(x) + 0.5
print(type(y))
print("123")
print(int(123.33))
print(int(float("123.33")))
```

### 3.字串的型別轉換

```python
x=10*3.25 # x=32.5
y=200*200 # y=4000
s="THE value of x is " + str(x) + ' and y is ' + str(y)

print(s)
print("The value of x is", x,"and y is",y)
print("The value of x is {} and y is {}".format(x,y))
```

## 非內建模組

```python
import math

print(math.pi)
print(math.pow(2,3))
print(8.3 * 10.58 * math.sin(37.0 / 180 * math.pi) / 2)
print(math.sqrt(25))
```

### 範例 : 算圓面積

```python
import math

r=input("請輸入圓的半徑:")
area=float(r) * float(r) * math.pi

print("圓面積為:",area)
```

### 應用題 : 算矩型面積

```python
r=input("請輸入矩形的長:")
w=input("請輸入矩形的寬:")
area=float(r) * float(w)

print("矩形面積為:",area)
```
